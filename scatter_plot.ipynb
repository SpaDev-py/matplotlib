```python
# Scatter plots are used to view relationships between two sets of values and see if they are correlated
# Scatter plots are really nice for seeing different trends and outliers e.t.c

import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn') # A stylesheet package is used to customize the matplotlib default properties



# creating random vaues of x and y
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
```

```python
# to view the viz using a scatter plot

plt.scatter(x,y);
```

```python
# To modify the sizes of the dots just add a 'size' parameter to the scatter function

plt.scatter(x,y, s = 100)  # s represents the size
```

```python
# changing the color

plt.scatter(x,y, s = 100, c = 'green'); # c represents the color
```

```python
# Changing the rounded dots to 'x'

plt.scatter(x,y, s= 100, c = 'green', marker = 'x')
```

```python
# adding more attributes

plt.scatter(x,y, s= 300, c = 'green', edgecolor = 'black', linewidth = 5,alpha = 0.75)

# edgecolor = adds border color to the markers
# linewidth = the edge color thickness
# alpha = modifies the marker color softness
```

```python
# Additonal data/information to the dataset, making this additional data correspond to the existing values

colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]

plt.scatter(x,y, s = 100, c = colors, edgecolor = 'black', linewidth = 1, alpha = 0.75)

# from the viz it more like shows the concentration of distribution of each data value
```

```python
# Instead of having shades of grey colors, we add a styling color to represent the intensity of each distribution
# by adding cmap (colormap)

plt.scatter(x,y, s = 100, c = colors, cmap = plt.cm.get_cmap('hot',5), edgecolor = 'black', linewidth = 1) #alpha = 0.75)

# the lighter 'Greens' are closer to zero while the darker ones are closer to 10 since 10 is somewhat the max scaling for the graph
# plt.cm.get_cmap('',<number>) --> a method from the cmap class that fetches a group of colors and applies them e.g 'hot', 'cold'



cbar = plt.colorbar(orientation = 'vertical', extend = 'both', pad = 0.05, shrink = 1, aspect = 20, format = '%.3f')
cbar.set_label('Satisfaction', size = 15)
cbar.set_ticks([0,5,10])
cbar.set_ticklabels(['babbs', 'spades', 'midas'])
cbar.ax.tick_params(labelsize = 15)
plt.clim(0,10)


# cbar is an object initialized to use functions and methods of the colorbar class
# orientation --> changes the view of the colorbar
# extend --> changes the colorbar shapes to arrow head 'both' indicating changes to be made to both ends
# pad --> padding of the colorbar to the graph border
# shrink --> shrinks the lenght of the colorbar, 1 makes it the same size as the graph, < 1 makes it smaller, > 1 makes it larger
# aspect --> manipulates the width of the colorbar making it wider or less. the smaller the number the wider.
# format --> changes the format of the numbers in terms of number of 0s. '%.3f' represents 3 0s.
# set_ticks --> sets the intervals for ticks on the colorbar to be placed
# set_ticklabels --> renames the ticks. customized mine because why not
# ax.tick_params(labelsize = 15) --> size of the tick labels
# cliim(0, 10) --> colorbar limits the max and min limits on the colorbar


# By adding the cmap the viz might be clear but not that clear as the colors dont really communicate anything
# so add a label to describe the cmap by using color bar, assuming for examplr the color represents customer satisfaction of a product
```
